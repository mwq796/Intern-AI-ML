### Plotting

```python
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=df)
plt.title("Sepal vs Petal length")
plt.show()
```

```python
sns.histplot(df['SepalLengthCm'], kde=True)
plt.title("Histogram of Petal_Length")
plt.show()
```

```python
sns.histplot(df['SepalWidthCm'], kde=True)
plt.title("Histogram of Petal_Width")
plt.show()
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=df)
plt.title("Petal Length by Species")
plt.show()
```

```python
sns.pairplot(df, hue="Species")
plt.show()
```
